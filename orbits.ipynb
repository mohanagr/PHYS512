```python
import numpy as np
import scipy.integrate as intg
import matplotlib.pyplot as plt
```

```python
def dYdt(U, th):
    # return the derivatives for given input
    #    500 km = 1.62*10^-14 kpc
    #    1 s = 3.18 10*-14 Myr

    l = 2 #Msun*kpc^2/Myr
    Vc = 2 #kpc/Myr
    
    u1 = U[0]
    u2 = U[1]
    u3 = U[2]
    
    u1_prime = u2
    
    u2_prime = -Vc**2/u1 + l**2/u1**3
    
    u3_prime = l/u1**2
    
    return [u1_prime, u2_prime, u3_prime]
```

```python
l = 2 #Msun*kpc^2/Myr
Vc = 2 #kpc/Myr

u1 = 2*l/Vc #r
u2 = 0
u3 = Vc**2/(4*l)

t = np.linspace(0,10,1000)
U0 = np.asarray([u1,u2,u3])
```

```python
Us = intg.odeint(dYdt, U0, t)
```

```python
x = Us[:,0]*np.cos(Us[:,2])
y = Us[:,0]*np.sin(Us[:,2])
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,8)
ax.set_title('Orbit for 10 Million years period')
ax.plot(x,y, 'r.', label='positions of star')

leg = ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
```
